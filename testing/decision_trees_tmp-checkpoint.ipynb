Predicting Salaries with Decision Trees
=======================================


![title](img/tree_5000_blank.png)

Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn import tree
```

```python
attribute_names = (
    'age',
    'workclass',
    'fnlwgt',
    'education',
    'education-num',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'capital-gain',
    'capital-loss',
    'hours-per-week',
    'native-country',
    'salary-class',
)
```

Read a CSV data file.

```python
df = pd.read_csv("adult.data", names=attribute_names, sep=', ', engine='python')
df.head()
```

Pre-process the data.  First, remove unwanted variables.

```python
excludes = (
    'fnlwgt', 
    'education-num',
    'capital-gain', 
    'capital-loss'
)
for att in excludes:
    del df[att]
df.head()
```

Boolean variables to 0s and 1s.

```python
booleans = (
    'sex', 
    'native-country', 
    'salary-class'
)
for att in booleans:
    df[att] = df[att].astype('category')
    df[att] = df[att].cat.codes
df.head()
```

 One-hot encoding of all categorical attributes.

```python
categoricals = (
    'workclass', 
    'education', 
    'marital-status', 
    'occupation', 
    'relationship', 
    'race'
)
pd.get_dummies(df['occupation'], prefix='occupation')

for att in categoricals:
    df = pd.concat([df, pd.get_dummies(df[att], prefix=att)], axis=1)
    del df[att]
df.head()
```

Extract attributes (X) and class labels (y).

```python
y = df['salary-class']
y.head()
```

```python
X = df.copy()
X = X.drop(columns='salary-class')
X.head()
```

Split dataset into training and testing subsets.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.33,
    random_state=0
)
```

Construct a decision tree from this training data.

```python
clf = DecisionTreeClassifier(
    random_state=0,
    criterion='entropy'
)
clf.fit(X_train, y_train)
```

Predict target class for training and testing sets, and check accuracy.

```python
y_train_pred = clf.predict(X_train)
train_acc = accuracy_score(y_train, y_train_pred)
print("Train: ", train_acc)

y_test_pred = clf.predict(X_test)
test_acc = accuracy_score(y_test, y_test_pred)
print("Test: ", test_acc)
```

Check if dataset is balanced.

```python
df['salary-class'].value_counts()
```

Prune the decision tree.

```python
min_split_sizes = [2,3,5,10,20,50,100,250,500,750,1000,2500,5000,10000,15000]
train_accs = []
test_accs = []
for s in min_split_sizes:
    clf = DecisionTreeClassifier(
        random_state=0,
        criterion='entropy',
        min_samples_split=s
    )
    clf.fit(X_train, y_train)   
    
    print("split size:", s)
    y_train_pred = clf.predict(X_train)
    train_acc = accuracy_score(y_train, y_train_pred)
    print("Train: ", train_acc)

    y_test_pred = clf.predict(X_test)
    test_acc = accuracy_score(y_test, y_test_pred)
    print("Test: ", test_acc)
    
    train_accs.append(train_acc)
    test_accs.append(test_acc)
plt.plot(min_split_sizes, train_accs)
plt.plot(min_split_sizes, test_accs)
plt.show()
```
